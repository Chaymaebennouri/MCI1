**Measure of central tendency**
(mode, mean, median)

- value spread of data
- Measures of Shape (Shape of distribution, skewness (to left or right = more values above or under mean), kurtosis (flatness or peakedness of symmetrical distribution))
- Measures of Concentration: Lorenz curve is not a statistic but the Gini index is (area between Lorenz curve and the first bisector=concentration area)




***Spread of Data***


*   Empirical Variance
*   Standard Deviation
*   Interquartile Ranges
*   Box and Whisker Plots


***Measures of Shape***

*   Shape of distribution
*   Skewness (to left or right = more values above or under mean)
*   Kurtosis (flatness or peakness of symmetrical distribution)


***Measures of Concentration***



*   Lorenz curve is not a statistic 

*   Gini index is the area between Lorenz curve and the first bisector = concentration area

```python
#Getting Percentage distribuition of "Potability" across the dataset
water_portability_df_counts = water_portability_df.groupby(['Potability'])['ph'].agg('count').reset_index(name="PH_scale")
fig1, ax1 = plt.subplots()
ax1.pie(water_portability_df_counts['PH_scale'],labels=water_portability_df_counts['Potability'],colors =["pink","blue"], explode= (0,0.05),autopct = '%1.1f%%')
ax1.set_title('Percentage of PH Scale across Portability',fontsize = 12)
```
